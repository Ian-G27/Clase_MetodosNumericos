# METODO DE NEVILLE

````python
import numpy as np
import pandas as pd
#Estos son los datos de entrada que utilizamos como referencia del libro
x = np.array([1.0, 1.3, 1.6, 1.9, 2.2])
f = np.array([0.7651977, 0.6200860, 0.4554022, 0.2818186, 0.1103623])
n = len(x)
# Se inicializara la tabla Q
Q = np.zeros((n, n))
Q[:, 0] = f
#Aplicare el algoritmo de Neville para construir la tabla
for i in range(1, n):
    for j in range(1, i + 1):
        Q[i, j] = ((x[i] - x[i - j]) * Q[i, j - 1] - (x[i] - x[i]) * Q[i - 1, j - 1]) / (x[i] - x[i - j])
#Voy a crear un dataframe con etiquetas de columnas apropiadas para simular una tabla igual a la del libro
columna_nombres = ["i", "xᵢ", "f[xᵢ]"] + [f"f[x{i - j},...,x{i}]" for i in range(1, n) for j in range(i)]
table_data = []
for i in range(n):
    row = [i, x[i], Q[i, 0]]
    for j in range(1, i + 1):
        row.append(Q[i, j])
    table_data.append(row)
#Ajustare nombres de las columnas reales
max_len = max(len(row) for row in table_data)
columna_nombres = columna_nombres[:3] + [f"f[x{i - j},...,x{i}]" for i in range(1, n) for j in range(i)]
columna_nombres = columna_nombres[:max_len]
# Por ultimo creare el dataframe
df = pd.DataFrame(table_data, columns=columna_nombres)
df.round(7) #Redondeare a 7 decimales para que sean los resultados similares al ejemplo del libro
````
