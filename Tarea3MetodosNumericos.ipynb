```python
#Librerias que seran utilizadas para el hacer el codigo del metodo de biseccion ⬇️
import sympy as sp
from tabulate import tabulate

#Primero voy a pedir al usuario que ingrese la funcion que desee⬇️
funcion_str = input("Ingresa la funcion f(x): ")   #La funcion se ingresa de la siguiente forma  x**3 + 4*x**2 - 10 (al querer elevar a una potencia se pondra "**")
#Creare la variable simbolica x de sympy⬇️
x = sp.symbols('x')
#Convertire la cadena de texto ingresada a la funcion simbolica usando sympy⬇️
f_simbolica = sp.sympify(funcion_str)
#Convertire la funcion simbolica en una funcion evaluable con lambdify esto convierte expresiones simbolicas en funciones numericas⬇️
f = sp.lambdify(x, f_simbolica)
#Ahora si definiendo el metodo de biseccion⬇️
def biseccion(a, b, TOL, N0):
    i = 1  #Inicializando el contador de iteraciones
    FA = f(a)  #Calculamos f(a)
    #Creare la lista para almacenar los valores a, b, p, f(p) en cada iteracion que se haga en la tabla⬇️
    tabla = []
    while i <= N0:
        p = (a + b) / 2  #Calculamos el punto medio p
        FP = f(p)  #Calculamos f(p)
        #Almacenamos los valores en la tabla para esta iteracion⬇️
        tabla.append([i, a, b, p, FP])
        #Verificamos si se ha encontrado una solucion exacta o si se cumple la tolerancia⬇️
        if FP == 0 or (b - a) / 2 < TOL:
            print("Solucion encontrada en la iteracion:", i)
            print(tabulate(tabla, headers=["n", "a", "b", "p", "f(pn)"]))
            return p

        i += 1  #Se incrementa el contador

        #Determinamos el siguiente intervalo de [a, b]⬇️
        if FA * FP > 0:
            a = p  #Actualizamos a
            FA = FP  #Actualizamos ahora f(a)
        else:
            b = p  #Actualizamos b

    #Por si no se encuentra la solucion despues de N0 iteraciones⬇️
    print(f"El metodo fracaso despues de {N0} iteraciones")
    print(tabulate(tabla, headers=["n", "an", "bn", "pn", "f(pn)"]))
    return None
#Pediendo al usuario que ingrese los valores necesarios que el desea ingresar⬇️
a = float(input("Ingresa el extremo izquierdo del intervalo (a): "))
b = float(input("Ingresa el extremo derecho del intervalo (b): "))
TOL = float(input("Ingresa la tolerancia (TOL): "))
N0 = int(input("Ingresa el numero maximo de iteraciones que quisieras (N0): "))
#Por ultimo se ejecutara el metodo de biseccion⬇️
solucion = biseccion(a, b, TOL, N0)
if solucion is not None:
    print(f"La solucion aproximada es: {solucion}")
```
