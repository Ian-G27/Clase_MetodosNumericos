# Diferencias Divididas de Newton

Paso 1: Calcula la tabla triangular de diferencias divididas (como en el pseudoc√≥digo).

Paso 2: Usa pandas para imprimir la tabla.

image.png


```python
import numpy as np
import pandas as pd
#Datos de entrada
x = np.array([1.0, 1.3, 1.6, 1.9, 2.2])
f = np.array([0.7651977, 0.6200860, 0.4554022, 0.2818186, 0.1103623])
n = len(x)
#Inicializa la matriz de diferencias divididas
F = np.zeros((n, n))
F[:, 0] = f  #Primera columna: f(xi)
#Paso 1:Se calculan las diferencias divididas
for i in range(1, n):
    for j in range(1, i + 1):
        F[i, j] = (F[i, j - 1] - F[i - 1, j - 1]) / (x[i] - x[i - j])
#Crear tabla en pandas para impresion
column_names = ["i", "x·µ¢", "F[i,0]"] + [f"F[i,{j}]" for j in range(1, n)]
table_data = []
for i in range(n):
    row = [i, x[i], F[i, 0]]
    for j in range(1, i + 1):
        row.append(F[i, j])
    table_data.append(row)
#Limitan las columnas seg√∫n longitud maxima de filas
max_len = max(len(row) for row in table_data)
column_names = column_names[:max_len]
df = pd.DataFrame(table_data, columns=column_names)
print("\nTabla de diferencias divididas de Newton:\n")
print(df.round(7))
```
