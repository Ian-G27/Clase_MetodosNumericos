# Metodo del punto fijo

````python
import numpy as np
from sympy import symbols, sympify, lambdify, N
from tabulate import tabulate

def iteracion_punto_fijo(g, p0, TOL, N0):
    tabla_resultados = []
    i = 0

    while i <= N0:
        p = g(p0)
        error = abs(p - p0)

        tabla_resultados.append([i, p0, p, error])

        if error < TOL:
            break
        p0 = p
        i += 1
    headers = ["n", "p_n", "g(p_n)", "Error"]
    print("\nTabla de resultados:")
    print(tabulate(tabla_resultados, headers=headers, tablefmt="grid", floatfmt=".10f"))

    if i > N0:
        print(f"\nEl metodo fallo despues de {N0} iteraciones.")
    else:
        print(f"\nLa raiz aproximada es: {p:.10f}")

def main():
    print("Metodo de Iteracion de Punto Fijo")
    print("---------------------------------")

    print("\nIngresa la funcion g(x):")
    func_str = input("g(x) = ")

    try:
        x = symbols('x')
        expr = sympify(func_str)
        g = lambdify(x, expr, 'numpy')
    except Exception as e:
        print(f"Ocurrio un error al analizar la funcion: {e}")
        return

    try:
        p0 = float(input("\nIngresa el valor inicial p0: "))
        TOL = float(input("Ingresa la tolerancia (TOL): "))
        N0 = int(input("Ingresa el numero maximo de iteraciones (N0): "))

        iteracion_punto_fijo(g, p0, TOL, N0)

    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"Error inesperado: {e}")

if __name__ == "__main__":
    main()
````

# Explicacion del codigo

Mtodo de iteracin de punto fijo para encontrar una solucin de p = g(p).
    
Parametros que se utilizaran en el codigo:

g   --> Funcion de iteracion

p0  --> Aproximacion inicial

TOL --> Tolerancia

N0  --> Numero maximo de iteraciones

def iteracion_punto_fijo(g, p0, TOL, N0):

    while i <= N0:

        #Nos ayuda a guardar los datos en la tabla

        if error < TOL:      #Si la diferencia entre iteraciones es menor que la tolerancia, el mtodo converge y se termina

        p0 = p  #Actualizando el punto p0 para la siguiente iteracion
        i += 1  #Incrementando el contador de las iteraciones
