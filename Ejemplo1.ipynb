```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y = x**2
plt.plot(x, y, label="y = x²", color="red")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de y = x²")
plt.legend()
plt.show()
```
